（12）用数据透视表，计算每个类型的酒店的评分人数总数量。

```python
pd.pivot_table(df,index=['类型'],values=['评分人数'],aggfunc={'评分人数':np.sum})
```

（13）用数据透视表，计算每个类型的酒店价格的均值和标准差
